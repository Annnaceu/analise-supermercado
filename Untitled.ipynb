```python
# Contar a frequência de cada item
item_counts = df['itemDescription'].value_counts()

# Mostrar os 10 itens mais comprados
print(item_counts.head(10))
```

```python
# Agrupar os dados por 'Member_number' e 'Date'
basket = df.groupby(['Member_number', 'Date'])['itemDescription'].apply(list).reset_index()

# Visualizar as primeiras transações
print(basket.head())
```

```python
from mlxtend.preprocessing import TransactionEncoder

# Codificar a lista de itens
encoder = TransactionEncoder()
onehot = encoder.fit(basket['itemDescription']).transform(basket['itemDescription'])
basket_encoded = pd.DataFrame(onehot, columns=encoder.columns_)

# Visualizar a matriz
print(basket_encoded.head())
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Calcular os itens frequentes
frequent_items = apriori(basket_encoded, min_support=0.01, use_colnames=True)

# Gerar as regras de associação
rules = association_rules(frequent_items, metric="confidence", min_threshold=0.5)

# Visualizar as regras
print(rules.head())
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Calcular os itens frequentes com um suporte mais baixo
frequent_items = apriori(basket_encoded, min_support=0.005, use_colnames=True)

# Gerar as regras de associação com uma confiança mínima mais baixa
rules = association_rules(frequent_items, metric="confidence", min_threshold=0.3)

# Visualizar as regras
print(rules.head())
```

```python
# Agrupar os dados por cliente e listar os itens comprados
customer_basket = df.groupby('Member_number')['itemDescription'].apply(list).reset_index()
print(customer_basket.head())
```

```python
# Calcular o número de itens diferentes comprados por cliente
customer_basket['num_items'] = customer_basket['itemDescription'].apply(lambda x: len(set(x)))
customer_basket['total_items'] = customer_basket['itemDescription'].apply(lambda x: len(x))

# Visualizar os dados
print(customer_basket[['Member_number', 'num_items', 'total_items']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gráfico de distribuição do número de itens
plt.figure(figsize=(10, 6))
sns.histplot(customer_basket['num_items'], bins=30, kde=True)
plt.title('Distribuição do Número de Itens Comprados por Cliente')
plt.xlabel('Número de Itens')
plt.ylabel('Frequência')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

X = customer_basket[['num_items', 'total_items']] 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

# Aplicar o método do cotovelo
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Visualizar o método do cotovelo
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss)
plt.title('Método do Cotovelo')
plt.xlabel('Número de Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Supondo que você escolheu K = 3
kmeans = KMeans(n_clusters=3, random_state=0)
customer_basket['cluster'] = kmeans.fit_predict(X_scaled)

# Visualizar a distribuição de clusters
sns.countplot(x='cluster', data=customer_basket)
plt.title('Distribuição de Clientes por Cluster')
plt.show()
```

```python
# Calcular a média de itens por cluster
cluster_analysis = customer_basket.groupby('cluster').agg({
    'num_items': 'mean',
    'total_items': 'mean'
}).reset_index()

print(cluster_analysis)
```
